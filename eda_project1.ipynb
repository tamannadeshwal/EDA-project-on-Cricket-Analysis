Feature Engineering

```python
#11. Create a new feature called "Boundary Percentage", which calculates the percentage of runs scored from fours and sixes.
 
df["Boundary_Runs"] = df["Fours"] * 4 + df["Sixes"] * 6
df["Boundary_Percentage"] = (df["Boundary_Runs"] / df["Runs_Scored"].replace(0, np.nan)) * 100
df
```

```python
#12. Create a feature called "Batting Impact Score" based on runs, strike rate, and average. 
df["Batting_Impact_Score"] = (df["Runs_Scored"] * df["Batting_Strike_Rate"] * df["Batting_Average"]).pow(1 / 3)  
df
```

```python
#13. Generate a metric to evaluate a player's all-round performance considering both batting and bowling stats. 
bat_norm = (df["Runs_Scored"] - df["Runs_Scored"].min()) / (df["Runs_Scored"].max() - df["Runs_Scored"].min())
bowl_norm = (df["Wickets_Taken"] - df["Wickets_Taken"].min()) / (df["Wickets_Taken"].max() - df["Wickets_Taken"].min())
df["All_Rounder_Index"] = (bat_norm + bowl_norm) / 2
df
```

```python
##14. Categorize players into Aggressive, Balanced, and Defensive based on their strike rate.
def label_style(s):
    if s >= 140:
        return "Aggressive"
    elif s >= 110:
        return "Balanced"
    else:
        return "Defensive"
df["Playing_Style"] = df["Batting_Strike_Rate"].apply(label_style)
df

 
```

```python
#15. Compute the year-over-year growth in performance metrics for each player. 
df.sort_values(["Player_Name", "Year"], inplace=True)
df["YoY_Runs_Growth"] = (df.groupby("Player_Name")["Runs_Scored"].pct_change() * 100)
df["YoY_Wkts_Growth"] = (df.groupby("Player_Name")["Wickets_Taken"].pct_change() * 100)
df
```
